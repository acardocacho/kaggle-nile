```python
def number_of_days(date):
    spray_times = spray['Date'].unique()
    spray_delta = [(date - x).days for x in spray_times]
    after_spray =  0 if all(i < 0 for i in spray_delta) else min(i for i in spray_delta if i>0)
    
    if after_spray <= 0:
        return 0
    else:
        return np.exp(-after_spray)    
```
